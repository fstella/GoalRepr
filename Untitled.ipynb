```python
import numpy as np

import params

from maps import PlaceRepresentation
from maps import GridRepresentation
from conns import Connections
from exper import Experiment

import matplotlib.pyplot as plt
```

```python
para=params
para.get_par()

BaseMap=PlaceRepresentation(para.par)
BaseMap.New_Map(para.par)
BaseMap.Generate_RateMap(para.par)
plt.imshow(BaseMap.RateMap[:,:,20])
plt.show()
plt.scatter(BaseMap.Place_Centers[:,0],BaseMap.Place_Centers[:,1])
plt.show()


GridMap=GridRepresentation(para.par)
GridMap.New_Map(para.par)

GridMap.Generate_RateMap(para.par)
plt.imshow(GridMap.RateMap[:,:,20])
plt.show()


plt.scatter(GridMap.Grid_Centers[:,0,0],GridMap.Grid_Centers[:,1,0])
plt.show()

```

```python
FF_Conn=Connections(para.par,BaseMap.Place_Centers.shape[0],GridMap.Grid_Centers.shape[0],'dist_placetogrid',para.par.Conn_norm_ff)
FF_Conn.Initialize(para.par,BaseMap.Place_Centers,GridMap.Grid_Centers)

RR_Conn=Connections(para.par,GridMap.Grid_Centers.shape[0],GridMap.Grid_Centers.shape[0],'dist_gridtogrid',para.par.Conn_norm_re)
RR_Conn.Initialize(para.par,GridMap.Grid_Centers,GridMap.Grid_Centers)

```

```python
Exp=Experiment('PlaceToGrid')
Exp.Define_Positions(para.par,'goal')
```

```python
Exp.SampleOutput(para.par,FF_Conn.W,BaseMap,RR_Conn.W,'reverberation')
plt.show()


for t in range(3):
    plt.subplot(1,3,t+1)
    AA=np.reshape(Exp.OutputEvol[:,[10+t*30],:],(-1,int(GridMap.N_ec**0.5),int(GridMap.N_ec**0.5)))
    plt.imshow(sum(AA,0))
    plt.colorbar()

plt.figure(figsize=(30,30))

plt.show()

AA=np.reshape(Exp.Output,(-1,int(GridMap.N_ec**0.5),int(GridMap.N_ec**0.5)))
plt.imshow(sum(AA,0))
plt.colorbar()
plt.show()

# AA=np.reshape(Exp.Output,(para.par.RateMap_resol,para.par.RateMap_resol,-1))
# plt.imshow(AA[:,:,20])
# plt.colorbar()
# plt.show()


BaseMap.Remap(para.par)
BaseMap.Generate_RateMap(para.par)
# # plt.imshow(BaseMap.RateMap[:,:,0])
# # plt.show()
# # plt.scatter(BaseMap.Place_Centers[:,0],BaseMap.Place_Centers[:,1])
# # plt.show()

Exp.SampleOutput(para.par,FF_Conn.W,BaseMap,RR_Conn.W,'reverberation')
plt.show()

for t in range(3):
    plt.subplot(1,3,t+1)
    AA=np.reshape(Exp.OutputEvol[:,[10+t*30],:],(-1,int(GridMap.N_ec**0.5),int(GridMap.N_ec**0.5)))
    plt.imshow(sum(AA,0))
    plt.colorbar()

plt.figure(figsize=(30,30))

plt.show()


AA=np.reshape(Exp.Output,(-1,int(GridMap.N_ec**0.5),int(GridMap.N_ec**0.5)))
plt.imshow(sum(AA,0))
plt.colorbar()
plt.show()

# AA=np.reshape(Exp.Output,(para.par.RateMap_resol,para.par.RateMap_resol,-1))
# plt.imshow(AA[:,:,20])
# plt.colorbar()
# plt.show()
```
